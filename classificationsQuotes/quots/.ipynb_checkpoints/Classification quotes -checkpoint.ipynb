```python
import numpy as np
import pandas as pd

from collections import OrderedDict

from bs4 import BeautifulSoup

from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import normalize

import nltk
import pymorphy2
from matplotlib import pyplot as plt
```

### Извлечение данных

```python
def getDF(file_name):
    with open(file_name) as ff:
        text = ff.read()
        soup = BeautifulSoup(text, 'html.parser')
        
        quots = []
        evals = []
        sings = {'+': 1, '-': -1, '0': 0}
        
        for sent in soup.find_all('sentence'):
            quot = sent.find_all('speech')[0].get_text().strip()
            evaluate = sent.find_all('evaluation')[0].get_text().strip()
            
            if evaluate not in sings:
                continue
            
            quots.append(quot)
            evals.append(sings[evaluate])
            
        return pd.DataFrame.from_dict({'quot': quots, 'eval': evals})
```

```python
DF_train = getDF("train/news_eval_train.xml")
print len(DF_train)
DF_train.head()
```

```python
DF_test = getDF("test/news_eval_test.xml")
print len(DF_test)
DF_test.head()
```

### Предобработка цитат

```python
import string
def delPunctuation(text):
    return ''.join([i for i in text if i not in string.punctuation.replace('-', '')])
```

```python
class NormalAnalyzer:
    __stemmer = pymorphy2.MorphAnalyzer()
    __cache = {}
    
    def norm(self, token):
        token = token.strip()
        
        norm = ''
        if token not in self.__cache:
    
            res = self.__stemmer.parse(token)
            norm = res[0].normal_form
            self.__cache[token] = norm
        else:
            norm = self.__cache[token]
        return norm
    
normal_analizer = NormalAnalyzer()
```

```python
def normalize(words):
    norm_words = []
    for word in words:
        norm_words.append(normal_analizer.norm(word))
    return norm_words
```

```python
def normalize_sent(sent):
    return ' '.join(normalize(nltk.tokenize.word_tokenize(delPunctuation(sent))))
```

```python
vectorizer = CountVectorizer()
# vectorizer.fit(map(normalize_sent, pd.concat([DF_train, DF_test])['quot']))
vectorizer.fit(map(normalize_sent, DF_train['quot']))
def createCorpus(DF):
    corp = map(normalize_sent, DF['quot'])
    return vectorizer.transform(corp)
```

### Построение векторов

#### Векторная модель

```python
corpTrain = createCorpus(DF_train)
vec_train = corpTrain
```

```python
corpTest = createCorpus(DF_test)
vec_test = corpTest
```

#### TF-IDF

```python
transformer = TfidfTransformer(sublinear_tf=True)
```

```python
tf_idf = transformer.fit(corpTrain)
```

```python
tf_idf_train = tf_idf.transform(corpTrain)
tf_idf_test = tf_idf.transform(corpTest)
```

#### Частотная модель

```python
prob_train = normalize(corpTrain, norm='l1', axis=1)
prob_test = normalize(corpTest, norm='l1', axis=1)
```

### Модели

```python
from sklearn.metrics import f1_score
from sklearn.model_selection import GridSearchCV
```

```python
def print_score_vec(model):
    pred_train = model.predict(vec_train)
    pred_test = model.predict(vec_test)
    train_score = f1_score(pred_train, DF_train['eval'], average='micro')
    test_score = f1_score(pred_test, DF_test['eval'], average='micro')
    print "Train score: {}".format(train_score)
    print "Test score: {}".format(test_score)
```

```python
def print_score_tf_idf(model):
    pred_train = model.predict(tf_idf_train)
    pred_test = model.predict(tf_idf_test)
    train_score = f1_score(pred_train, DF_train['eval'], average='micro')
    test_score = f1_score(pred_test, DF_test['eval'], average='micro')
    print "Train score: {}".format(train_score)
    print "Test score: {}".format(test_score)
```

```python
def print_score_prob(model):
    pred_train = model.predict(prob_train)
    pred_test = model.predict(prob_test)
    train_score = f1_score(pred_train, DF_train['eval'], average='micro')
    test_score = f1_score(pred_test, DF_test['eval'], average='micro')
    print "Train score: {}".format(train_score)
    print "Test score: {}".format(test_score)
```

#### Naive bayes

```python
from sklearn.naive_bayes import MultinomialNB
```

Vector

```python
params = {'fit_prior': (True, False), 'alpha': np.arange(0.1, 2, 0.05)}

model = MultinomialNB()
grCV = GridSearchCV(model, params, scoring='f1_micro', cv=3)
grCV.fit(vec_train, DF_train['eval'])

means = grCV.cv_results_['mean_test_score']

plt.plot(np.arange(0.1, 2, 0.05), means[::2], label='fit_prior = True')
plt.plot(np.arange(0.1, 2, 0.05), means[1::2], label='fit_prior = False')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()

print_score_vec(grCV)
```

Tf-idf

```python
params = {'fit_prior': (True, False), 'alpha': np.arange(0.1, 2, 0.05)}

model = MultinomialNB()
grCV = GridSearchCV(model, params, scoring='f1_micro', cv=3)
grCV.fit(tf_idf_train, DF_train['eval'])

means = grCV.cv_results_['mean_test_score']

plt.plot(np.arange(0.1, 2, 0.05), means[::2], label='fit_prior = True')
plt.plot(np.arange(0.1, 2, 0.05), means[1::2], label='fit_prior = False')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()

print_score_tf_idf(grCV)
```
